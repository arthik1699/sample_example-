```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.sample(10)
```

```python
my_countries = ["USA", "IND", "CHN"]

df2 = df[df["Country Code"].isin(my_countries)]

df2.shape
```

```python
df2.head()
```

```python
df2.tail()
```

```python
df3 = df2.query('indicator == "SP.POP.TOTL" and Year == 2021')
df3
```

```python
fig = px.bar(df3, 
             x = "Country Name", 
             y ="value", 
             title = "2021 Population of USA, China and India", 
             color = "Country Name", 
             template = "plotly_dark")
fig.update_layout(showlegend = False)
fig.show()
```

```python
fig = px.pie(df3, 
             names = "Country Name", 
             values ="value", 
             title = "2021 Population of USA, China and India", 
             color = "Country Name", 
             template = "plotly_dark")
#fig.update_layout(showlegend = True)
fig.update_traces
fig.show()
```
